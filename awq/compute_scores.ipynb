```python
import numpy as np
import json
import os
import sys
sys.path.append('..')

from scoring_pipeline import ScoringPipeline
```

```python
sp = ScoringPipeline()
```

```python
base_dir = '/fs/cfar-projects/low-bit-vision/awq_results/image_text_retrieval'

results_dir = os.path.join(base_dir, 'image_text_retrieval_results')
scores_dir = os.path.join(base_dir, 'image_text_retrieval_scores')

os.makedirs(scores_dir, exist_ok=True)
```

```python
for i in range(7**2):
    result_path = os.path.join(results_dir, f'awq_{i}')

    scores = {}
    with open(result_path) as f:
        result = json.load(f)

        # some json post processing to get scoringpipeline to work
        result['scores_i2t'] = np.array(result['scores_i2t'])
        result['scores_t2i'] = np.array(result['scores_t2i'])
        result['txt2img'] = {int(k):v for k,v in result['txt2img'].items()}
        result['img2txt'] = {int(k):v for k,v in result['img2txt'].items()}

        print(f'Scoring result awq_{i}')
        scores = sp.compute_scores(result, 'image_text_retrieval')
        scores['model_size'] = result['model_size']
    

    # print(scores)
    scores_path = os.path.join(scores_dir, f'awq_{i}')
    with open(scores_path, 'w') as f:
        json.dump(scores, f, indent=2)
```
