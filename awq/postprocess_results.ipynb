```python
import json
import os
import pandas as pd
```

```python
awq_results_path = "./awq_results/image_text_retrieval"
```

```python
gather = []
for i in range(7**2):
    config_path = os.path.join(awq_results_path, "image_text_retrieval_configs", f"awq_{i}")
    config = json.load(open(config_path))

    scores_path = os.path.join(awq_results_path, "image_text_retrieval_scores", f"awq_{i}")
    scores = json.load(open(scores_path))

    # merge dicts :O
    row = config | scores
    gather.append(row)
```

```python
df = pd.DataFrame(gather)
df["vit_bits"] = [x["self_attn"] for x in df["vit_layers"]]
df["qformer_bits"] = [x["self_attn"] for x in df["qformer_layers"]]
df = df.drop(["vit_layers", "qformer_layers"], axis=1)
cols_to_move = ["vit_bits", "qformer_bits"]
df = df[cols_to_move + [col for col in df.columns if col not in cols_to_move]]
df[:5]
```

```python
df.to_csv(os.path.join(awq_results_path, "awq_image_text_retrieval.csv"), index=None)
```

```python
awq_results_path = "./awq_results/image_captioning"
```
